```python
from numpy import genfromtxt
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# 数据读取
abalone = pd.read_csv('/Users/wangmurong/Projects/math5836/abalone/abalone.data')
abalone.columns = ['Sex' ,'Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight','Shell weight','Rings']
# print(abalone.head(5))

# 数据预处理
abalone['Sex'][abalone['Sex']=='M']=1
abalone['Sex'][abalone['Sex']=='F']=0
abalone['Sex'][abalone['Sex']=='I']=2
# df.apply(pd.to_numeric, errors='ignore')
print(abalone['Sex'].dtype)
# abalone['Sex'] = abalone['Sex'].astype(str).astype()
abalone['Sex'] = pd.to_numeric(abalone['Sex'], errors='coerce')
print(abalone['Sex'].dtypes)

# 相关性图绘制
correlation_matrix = abalone.corr().round(2)
sns.heatmap(data=correlation_matrix,annot=True)


# 散点图绘制
features = ['Sex', 'Shell weight']
target = abalone['Rings']
for i, col in enumerate(features): 
    plt.subplot(1, len(features) , i+1) 
    x = abalone[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('Rings')
    plt.savefig('feature.png')

plt.close()

# abalone['Shell Weight'].plot()
# # 直方图绘制
plt.hist(abalone['Shell weight'],bins = 10,
         alpha = 0.45, color = 'red')
plt.hist(abalone['Sex'], bins=10, alpha=0.45, color='blue')
plt.show()
# # abalone[['Rings','Sex','Shell weight']].plot.hist(bins=3)
# # abalone['Rings'].hist(by=abalone['Shell weight'])
X = abalone[['Sex' ,'Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight','Shell weight']]
Y = abalone['Rings']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.4, random_state=5)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score 
import numpy as np

# 模型训练
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
# model evaluation for training set
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict))) 
r2 = r2_score(Y_train, y_train_predict)
print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

# 模型测试
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict))) 
r2 = r2_score(Y_test, y_test_predict)
print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
